```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
CovidData = pd.read_csv("All_DataSets/covid19_Confirmed_dataset.csv")
CovidDeathData = pd.read_csv("All_DataSets/covid19_deaths_dataset.csv")
CovidHappines = pd.read_csv("All_DataSets/worldwide_happiness_report.csv")
```

```python
#Wehave To Delete Some Unwanted Columns To make It Useful
#Then We will Transpose The dataframe in such way that country will be column and dates will be an another column
#The Columns Province/State Lat Long All this columns Has no use So we Delet It
CovidData.drop(["Province/State","Lat","Long"] ,inplace = True ,axis = 1)
```

```python

# Step 1: Melt (unpivot dates into rows)
df = CovidData.melt(
    id_vars=["Country/Region"],   
    value_vars=CovidData.columns[4:],   
    var_name="Date",
    value_name="Cases"
)

# Step 2: Convert Date to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Step 3: Aggregate (sum over provinces)
df_grouped = df.groupby(["Date", "Country/Region"], as_index=False)["Cases"].sum()

# Step 4: Pivot (countries as columns)
new_df = df_grouped.pivot(
    index="Date",
    columns="Country/Region",
    values="Cases"
).reset_index()

new_df.head(50)
```
